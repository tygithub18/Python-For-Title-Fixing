```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Bring in Pandas, Numpy, and matplotlib
#These are libraries we are bringing in to use later
```

```python
data = pd.read_csv("Title List.csv")
#open the file
```

```python
data
#Here is our raw data. It is small, so it is ok to display.
```

```python
data['Len name'] = data['Extracted Title Name'].str.len()
'''I need to get the length of the titles, there are head and tail
blank spaces and tabs to remove.'''
#Another way to do this is below, better if no spaces in name
#data['Len name'] = data.Extracted English name.apply(len)
```

```python
data.head()
#See the first row, it should be a length of 10
```

```python
data['Extracted Title Name'] = data['Extracted Title Name'].str.strip()
#Remove the extra spaces and tabs
```

```python
data['Len name'] = data['Extracted Title Name'].str.len()
#Re-count the length
```

```python
data.head()
#You can see that I removed the extra tabs and spaces
```

```python
data = data.drop("Len name", axis=1)
#I don't need this column anymore. so drop it
```

```python
data.head()
#See the "Len name" column is gone.
```

```python
data['Title name'] = data['Extracted Title Name'].str.title()
#Time to fix the names. Excel has the fx "proper"
```

```python
data.tail()
#While it did fix titles,the problem is that str.title() will capitliaze the 's in the title.
#Look at Char's
```

```python
data['Title name'] = data['Title name'].str.replace("'S","'s")
data['Title name'] = data['Title name'].str.replace("Part Ii","'Part II")
data['Title name'] = data['Title name'].str.replace("Part IIi","'Part III")
#Change the capital S to a lowered s
#Fix Part 2 and 3. 3 is just in case.
```

```python
data.tail()
#Now it is fixed
```

```python
data['Original Title name'] = data['Title name']
#I want to use this to see which records I updated later on
```

```python
data['Last chars']= data['Title name'].str[-5:]
#This part is for ", the" it is 5 characters at the end.
```

```python
data.head()
#See the last 5 characters.
```

```python
data.head()
#See, even a 4 char word is ok.
```

```python
data['The prob1']= np.where(data['Last chars']==', The','Yes','No')
#I want to know which columns are equal to ', The'
```

```python
data.tail()
#See the, Yes
```

```python
data['Fix Title name1']= np.where(data['The prob1']== 'Yes', 'The '+data['Title name'],data['Title name'])
#Now I am appending 'The ' in the front of the ones with the ', The' problem.
```

```python
data.tail()
#See row 30 with Dark Knight
```

```python
data['Title name']= np.where(data['The prob1']== 'Yes', data['Fix Title name1'].str[:-5],data['Fix Title name1'])
#Now I am going to fix/paste over the Title name column with the proper name.
#Take off the last 5 charscters from the Fix Title name1 and that becomes the Title name.
```

```python
data.tail()
#See the tiles in Title name have been fixed.
```

```python
#This is the same code compressed together.
#For ',The' cases
data['Last chars']= data['Title name'].str[-4:]
data['The prob1']= np.where(data['Last chars']==',The','Yes','No')
data['Fix Title name1']= np.where(data['The prob1']== 'Yes', 'The '+data['Title name'],data['Title name'])
data['Title name']= np.where(data['The prob1']== 'Yes', data['Fix Title name1'].str[:-4],data['Fix Title name1'])
```

```python
data.tail()
```

```python
#For the '/The'
data['Last chars']= data['Title name'].str[-4:]
data['The prob1']= np.where(data['Last chars']=='/The','Yes','No')
data['Fix Title name1']= np.where(data['The prob1']== 'Yes', 'The '+data['Title name'],data['Title name'])
data['Title name']= np.where(data['The prob1']== 'Yes', data['Fix Title name1'].str[:-4],data['Fix Title name1'])
```

```python
#For the ', A'
data['Last chars']= data['Title name'].str[-3:]
data['The prob1']= np.where(data['Last chars']==', A','Yes','No')
data['Fix Title name1']= np.where(data['The prob1']== 'Yes', 'A '+data['Title name'],data['Title name'])
data['Title name']= np.where(data['The prob1']== 'Yes', data['Fix Title name1'].str[:-3],data['Fix Title name1'])
```

```python
#For the ',A'
data['Last chars']= data['Title name'].str[-2:]
data['The prob1']= np.where(data['Last chars']==',A','Yes','No')
data['Fix Title name1']= np.where(data['The prob1']== 'Yes', 'A '+data['Title name'],data['Title name'])
data['Title name']= np.where(data['The prob1']== 'Yes', data['Fix Title name1'].str[:-2],data['Fix Title name1'])
```

```python
data.tail(15)
#Do I need to re proper the casing again? No
```

```python
data['Title Updated?']= np.where(data['Original Title name'] != data['Title name'],'Yes','No')
```

```python
data.tail()
```

```python
data = data.drop(['The prob1', 'Fix Title name1','Last chars', 'Original Title name'], axis=1)
#Now we can get rid of the extra columns
```

```python
data
#Here is our finished clean data
```

```python
data.loc[data['Title Updated?'] == 'Yes']
#Let's see only the update titles
```

```python
#Now I want to know if a title ends with an close parenthesis this is probably bad
#But first I need to take care of the ',a' and ', A'
#Also ones that end with a number. Might be a year number reference
```

```python
data['Ends with )'] = data['Title name']
```

```python
data['Ends with )'] = data['Ends with )'].str[-1]
```

```python
data['Ends with )'] = np.where(data['Ends with )'] == ')' , 'Yes', 'No')
```

```python
data.tail()
```

```python
data['Ends_with_num'] = data['Title name']
```

```python
data['Ends_with_num'] = data['Ends_with_num'].str[-1]
```

```python
data['Ends_with_a_num'] = data.Ends_with_num.map({'0':"Is_num", '1':"Is_num", '2':"Is_num", '3':"Is_num", '4':"Is_num", '5':"Is_num", '6':"Is_num", '7':"Is_num", '8':"Is_num", '9':"Is_num",})
#I used map to apply a new column based on an exisitings columns values
```

```python
data = data.drop(['Ends_with_num'], axis=1)
#I don't need this column anymore.
```

```python
#Find any title that says to use something else, like a different MPM#
data['Use_Something_Else'] = data['Title name'].str.contains("- Use ")
```

```python
data.tail()
```

```python
#11.9.18 I want to create a status column so that I can use it for a dashboard.
#I can use it for a bar or pie chart to show the % of the data that is clean, was cleaned, and is questionable
data['Status'] = np.where(data['Title Updated?'] == 'Yes' , 'Cleaned up', 'Clean')
data['Status'] = np.where(data['Ends with )'] == 'Yes' , 'Questionable', data['Status'])
data['Status'] = np.where(data['Ends_with_a_num'] == 'Is_num' , 'Questionable', data['Status'])
data['Status'] = np.where(data['Use_Something_Else'] == True , 'Questionable', data['Status'])
```
